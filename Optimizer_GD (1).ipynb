```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv('/content/diabetes.csv')
```

```python
dataset['Glucose']=np.where(dataset['Glucose']==0,dataset['Glucose'].median(),dataset['Glucose'])
dataset['Insulin']=np.where(dataset['Insulin']==0,dataset['Insulin'].median(),dataset['Insulin'])
dataset['BloodPressure']=np.where(dataset['BloodPressure']==0,dataset['BloodPressure'].median(),dataset['BloodPressure'])
dataset['SkinThickness']=np.where(dataset['SkinThickness']==0,dataset['SkinThickness'].median(),dataset['SkinThickness'])
dataset['BMI']=np.where(dataset['BMI']==0,dataset['BMI'].median(),dataset['BMI'])
```

```python
dataset.describe()
```

```python
x=dataset.iloc[:,0:-1]
```

```python
y=dataset.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
#splitting
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_scaled=sc.fit_transform(x_train)
x_test_scaled=sc.transform(x_test)
```

```python
x_train_scaled
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
import time
```

```python
#Stochastic nature
model=Sequential()
model.add(Dense(32,activation='relu',input_dim=8))
model.add(BatchNormalization())
model.add(Dropout(0.25))
model.add(Dense(16,activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='sgd',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
start=time.time()
model.fit(x_train_scaled,y_train,epochs=10,batch_size=1,validation_data=(x_test_scaled,y_test))
end=time.time()
```

```python
print(end-start)
```

```python
#Batch nature
model_1=Sequential()
model_1.add(Dense(32,activation='relu',input_dim=8))
model_1.add(BatchNormalization())
model_1.add(Dropout(0.25))
model_1.add(Dense(16,activation='relu'))
model_1.add(Dropout(0.25))
model_1.add(Dense(1,activation='sigmoid'))
```

```python
model_1.compile(optimizer='sgd',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
start=time.time()
model_1.fit(x_train_scaled,y_train,epochs=10,batch_size=614,validation_data=(x_test_scaled,y_test))
end=time.time()
```

```python
print(end-start)
```
