```python
digits=load_digits()
```

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
```

```python
x=digits['data']
```

```python
y=digits['target']
```

```python
pd.DataFrame(x).shape
```

```python
pd.DataFrame(y).shape
```

```python
pd.DataFrame(y).value_counts()
```

```python
pd.DataFrame(y).nunique()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
p=Perceptron()
```

```python
p.fit(x_train,y_train)
```

```python
prediction=p.predict(x_test)
```

```python
prediction
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,prediction)
```

```python
prediction_training=p.predict(x_train)
```

```python
accuracy_score(y_train,prediction_training)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,prediction))
```

```python
print(classification_report(y_train,prediction_training))
```

```python
dataset=pd.read_csv('/content/HousingData.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset.isnull().sum()/len(dataset)*100
```

```python
x=dataset.iloc[:,0:-1]
```

```python
y=dataset.iloc[:,-1]
```

```python
x.shape
```

```python
y.shape
```

```python
x_train_initial,x_test_initial,y_train_initial,y_test_initial=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train_initial.shape)
print(x_test_initial.shape)
print(y_train_initial.shape)
print(y_test_initial.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train_initial_scaled=sc.fit_transform(x_train_initial)
```

```python
x_test_initial_scaled=sc.transform(x_test_initial)
```

```python
pd.DataFrame(x_train_initial_scaled)
```

```python
pd.DataFrame(x_test_initial_scaled)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model_without_imputation=Sequential()
model_without_imputation.add(Dense(32,activation='relu',input_dim=13))
model_without_imputation.add(Dense(16,activation='relu'))
model_without_imputation.add(Dense(1,activation='linear'))
```

```python
model_without_imputation.summary()
```

```python
model_without_imputation.compile(optimizer='adam',loss='mse')
```

```python
model_without_imputation.fit(x_train_initial_scaled,y_train_initial,epochs=100,batch_size=32,validation_data=(x_test_initial_scaled,
                                                                                                              y_test_initial))
```

```python
from sklearn.impute import KNNImputer
```

```python
x_imputed=imputer.fit_transform(x)
```

```python
print(x_train.shape)
print(x_testl.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x_train_scaled=sc.fit_transform(x_train)
```

```python
x_test_scaled=sc.transform(x_testl)
```

```python
x_train_scaled
```

```python
x_test_scaled
```

```python
model_with_imputation=Sequential()
model_with_imputation.add(Dense(32,activation='relu',input_dim=13))
model_with_imputation.add(Dense(16,activation='relu'))
model_with_imputation.add(Dense(1,activation='linear'))
```

```python
model_with_imputation.summary()
```

```python
model_with_imputation.compile(optimizer='adam',loss='mse')
```

```python
model_with_imputation.fit(x_train_scaled,y_train,epochs=100,batch_size=32,validation_data=(x_test_scaled,y_test))
```
